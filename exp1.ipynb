```python
from PIL import Image
img_path = 'hanu.jpg'
image_pil = Image.open(img_path)
save_path = 'testing.jpg'
image_pil.save(save_path)
image_pil.show()

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img_path = 'hanu.jpg'
image_matplotlib = mpimg.imread(img_path)
save_path = 'image_matplotlib.jpg'
plt.imsave(save_path, image_matplotlib)
plt.imshow(image_matplotlib)
plt.show()

from skimage import io
img_path = 'hanu.jpg'
image_skimage = io.imread(img_path)
save_path = 'image_skimage.jpg'
io.imsave(save_path, image_skimage)
io.imshow(image_skimage)
io.show()
```
